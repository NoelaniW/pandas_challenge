```python
import pandas as pd
from pathlib import Path
```

```python
school_data_to_load = Path("Resources", "schools_complete.csv")
student_data_to_load = Path("Resources", "students_complete.csv")
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#DISTRICT SUMMARY
```

```python
#calculate total number of unique schools
```

```python
unique_schools = school_data_complete["school_name"].unique()
unique_schools
```

```python
unique_schools = school_data_complete["school_name"].nunique()
unique_schools
```

```python
#calculate total number of students
```

```python
student_count = len(school_data_complete["student_name"])
student_count
```

```python
#calculate total budget
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
#calculate average math score
```

```python
average_math_score = student_data["math_score"].mean()
average_math_score
```

```python
#calculate average reading score
average_reading_score = student_data["reading_score"].mean()
average_reading_score
```

```python
#calculate passing math percentage
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
#calculate passing reading percentage
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
district_summary = pd.DataFrame({"Total Schools": [unique_schools], 
                                 "Total Students": [student_count], 
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [average_math_score], 
                                 "Average Reading Score": [average_reading_score],
                                 "Percent Passing Math": [passing_math_percentage],
                                 "Percent Passing Reading": [passing_reading_percentage],
                                 "Percent Overall Passing": [overall_passing_rate]})
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```

```python
#SCHOOL SUMMARY
```

```python
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
#calculate total number of students per school
per_school_counts = school_data.groupby("school_name")["size"].sum()
per_school_counts
```

```python
#calculate total budgets per school
per_school_budget = school_data.groupby("school_name")["budget"].sum()
per_school_budget
```

```python
#calculate average budget per student per school
per_school_capita = per_school_budget/per_school_counts
per_school_capita
```

```python
#calculate average math_score per school
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_math
```

```python
#calculate average reading score per school
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
per_school_reading
```

```python
#calculate number of students passing math per school
students_passing_math = school_data_complete[(school_data_complete["math_score"] >=70)]
students_passing_math
```

```python
school_students_passing_math = students_passing_math.groupby(["school_name"]).size()
school_students_passing_math
```

```python
#students passing reading per school
students_passing_reading = school_data_complete[(school_data_complete["reading_score"] >=70)]
students_passing_reading
```

```python
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size()
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100

per_school_passing_math
```

```python
per_school_passing_reading
```

```python
overall_passing_rate
```

```python
#Create dataframe summarize above calculations
per_school_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students per School": per_school_counts,
                                   "Budget Per School": per_school_budget,
                                   "Budget Per Student": per_school_capita,
                                   "Average Math Score": per_school_math,
                                   "Average Reading Score": per_school_reading,
                                   "Total Passing Math": per_school_passing_math,
                                   "Total Passing Reading": per_school_passing_reading,
                                   "Overall Passing Rate": overall_passing_rate})
per_school_summary["Budget Per School"] = per_school_summary["Budget Per School"].map("${:,.2f}".format)
per_school_summary["Budget Per Student"] = per_school_summary["Budget Per Student"].map("${:,.2f}".format)

per_school_summary
```

```python
#HIGHEST PERFORMING SCHOOLS (BY % OVERALL PASSING)
```

```python
#Sort % overall passing schools by descending order
top_schools = per_school_summary.sort_values("Overall Passing Rate", ascending=False)
top_schools.head(5)
```

```python
#Sort % overall passing schools by ascending order
bottom_schools = per_school_summary.sort_values("Overall Passing Rate", ascending=True)
bottom_schools.head(5)
```

```python
#SCORES BY SCHOOL SPENDING
```

```python
#Establish bins
spending_bins = [0, 585, 630, 645, 680]
bin_name = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary.copy()
```

```python
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=bin_name)
school_spending_df.head(5)
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Total Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Total Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing Rate"].mean()
```

```python
#Create dataframe for spending summary
spending_summary = school_spending_df.groupby("Spending Ranges (Per Student)")

spending_summary = spending_summary[["Average Math Score", "Average Reading Score", "Total Passing Math", 
                               "Total Passing Reading", "Overall Passing Rate"]]
spending_summary.mean()
```

```python
#SCORES BY SCHOOL SIZE
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
label_size = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students per School"], size_bins, labels=label_size)
per_school_summary.head()
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["Total Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["Total Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Overall Passing Rate"].mean()
```

```python
size_summary = per_school_summary.groupby("School Size")

size_summary = size_summary[["Average Math Score", "Average Reading Score", "Total Passing Math", 
                               "Total Passing Reading", "Overall Passing Rate"]]
size_summary.mean()
```
